مرحله چهارم: Mean Squared Error را برای پاسخ به دست آمده محاسبه کنید.
<br/><br/>
مرجله پنجم: تابع predict را متناسب با جواب به دست آمده تکمیل کنید، درستی جواب به دست آمده را می‌توانید با بررسی test case‌هایی که در انتها آمدهآند مشخص کنید.
<br/>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
 data_set = pd.read_csv('./dataset/house_dataset.csv')
print ("Rows num: ", data_set.size, "\n")

area = data_set.loc[0:]['Area']
price = data_set.loc[0:]['Price']
```

```python
# creating matrix A using vstack to stack array vertically (row - wise) and then transpose it work correctly
A = np.vstack([area, np.ones(len(area))]).T

# finding the line of answer by solving the least square of Ax = b
x, residuals, rank, s = np.linalg.lstsq(A, price, rcond=None)

# finding the slope and the width from orgin of our line
slope, y_intercept = x

# print(f"The slope of the line is {slope:.2f} and the y-intercept is {y_intercept:.2f}.")
```

```python
w = y_intercept
b = slope
print("w: ", w, " and b: ", b)
```

<div style="text-align: right">
مرحله سوم: نمایش پاسخ
</div>

```python
# showing the line (we can't show it using multy dimentional so we need some numpy arrays)
plt.plot(np.array(area), np.array(price), 'o', label='Original data', markersize=5)
plt.plot(np.array(area), slope*np.array(area) + y_intercept, 'r', label='Line')
plt.legend()
plt.show()
```

<div style="text-align: right">
مرحله چهارم : محاسبه Mean Squared Error
</div>

```python
# we know that mse = 1/n(sum of(y - yhat)^2)
predicted_price = slope*area + y_intercept
mse = np.mean((price - predicted_price)**2)

print("MSE: ", mse)
```

<div style="text-align: right">
مرحله پنجم : تکمیل تابع predict و پیش بینی قیمت خانه‌ها
</div>

```python
def predict(x) :
    # x is the area of our house , if we want to predict the price we need to put it on the line that we have found
    # and we know that price = slope*area + width from origin so : 
    return slope*x + y_intercept
```
